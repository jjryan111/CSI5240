```python
from matplotlib.axes import Axes
import pandas as pd
import re, string, random
import os
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import matplotlib.dates as dates
```

```python
df6=pd.read_csv('/Users/jj/PyProjects/CSI4240/monday.com_hack_data/df.csv')
for i in range(len(df6)):
    df6.at[i,'days'] = datetime.strptime(df6.at[i,'d'], '%Y-%m-%d')
        
```

```python
x_dates = df6['days']
x_num = dates.date2num(x_dates)
trend = np.polyfit(x_num, df6['value'], 1)
fit = np.poly1d(trend)
df6.plot(x='days', y='value')
plt.suptitle('Tweet Positivity Over Time: Monday.com', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Positive Tweet %', fontsize=16)
plt.plot(x_num, fit(x_num), "r--")
plt.savefig('/Users/jj/PyProjects/CSI4240/monday.com_hack_data/data1_chart.jpg')

plt.show()
```
